Now we will try to create two columns one with date and time and other with users name and mesasage

Here we can create a regular expression for our input either manually or using some of the exisiting tools like regex101.com

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

d{1,n} means digits can one to n number of digits
/ means talking about next one

```python
messages = re.split(pattern, data)
messages
```

```python
#let's remove that empty string
messages = re.split(pattern, data)[1:]
messages
```

```python
dates = re.findall(pattern, data)
dates
```

```python
import pandas as pd

df = pd.DataFrame({"user_message": messages, "message_date": dates})
df["message_date"] = pd.to_datetime(df["message_date"], format="%d/%m/%y, %H:%M - ")
df.rename(columns={"message_date": "date"}, inplace=True)
df.head()
```

```python
df.shape
```

Now Let's separate user name and message

```python
users=[]
messages=[]
for message in df["user_message"]:
    entry = re.split("([\w\W]+?):\s",message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append("group_notification")
        messages.append(entry[0])

df["user"]=users
df["messages"]=messages
df.drop(columns=["user_message"],inplace=True)
df.head()
```
